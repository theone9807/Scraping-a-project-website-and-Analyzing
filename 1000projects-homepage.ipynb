```python
import pandas as pd
import requests
from bs4 import BeautifulSoup as bs
df = pd.read_csv('1000projectsCatagory.csv')
li = df['CatagoryUrl'].tolist()
#list of catagory for each link
cat_list = []

def cat_finder(url):
    page = requests.get(url)
    soup = bs(page.content, 'html.parser')
    cat = soup.find(class_='cat-links')
    cat1 = cat.findAll('a')
    #appending multiple category to one row
    s = ''
    for c in cat1:
        s += str(c.get_text()) + ', '
        #print(s)

    #appending as single row
    cat_list.append(s)

# url = ['http://1000projects.org/android-mobile-application-for-booking-vehicle.html','http://1000projects.org/1000-projects.html']
for u in li[:2]:
    try:
        cat_finder(u)
    except:
        pass
    

dfbody = pd.DataFrame(cat_list, columns=["Cateogory"])
dfbody.to_csv('1000projectscataogryName.csv', index=False)
```
