```python
import requests
from bs4 import BeautifulSoup as bs

url = 'http://asokatechnologies.in/project-titles-for-b-tech-and-m-tech/'
page = requests.get(url)
soup = bs(page.content, 'html.parser')
e = soup.findAll(class_= 'entry-content')
for b in e:
    a = b.findAll('td')
print(soup)
```

```python
#a
body = []
import re
for b in a:
    if re.match(r'\w+',b.get_text()):
        if not b.get_text().isdigit(): 
            if b.get_text() != 'Download' and b.get_text() != 'Download ':
#                 print(b.get_text())
                body.append(b.get_text())
    
body
```

```python
urls = []
for e in a:
    urls.append(e['href'])
    
del urls[0]
urls = list(set(urls))
urls
```

```python
def scrapy(url):
    
#     url = 'http://asokatechnologies.in/single-phase-grid-connected/'
    page = requests.get(url)
    soup = bs(page.content, 'html.parser')
    e = soup.findAll(class_= 'entry-content')
    body = []
    for e in e:
        if e:
            body.append(e.get_text().replace('\n', ' ').replace('\t', '').strip())
        else:
            body.append('NA')
                
    return body[0]
```

```python
main = []
for url in urls:
    try:
        main.append(scrapy(url))
    except:
        pass
    
main
```

```python
import pandas as pd

df1 = pd.DataFrame(urls, columns=['Urls'])
df2 = pd.DataFrame(main, columns=['Body'])
df1.reset_index(drop=True, inplace=True)
df2.reset_index(drop=True, inplace=True)
df = pd.concat([df1, df2], axis=1)
```
